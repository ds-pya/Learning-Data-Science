### 2) 데이터 표현

#### 위치
데이터를 살펴보는 가장 기초적인 단계는 각 변수의 대푯값을 구하는것.  
> 평균 / 가중평균 / 중간값 / 가중 중간값 / 절사평균 등

<br>

#### 변이
이후 위에서 데이터의 '위치'를 살펴보았다면, '변이'도 살펴봐야함  
> 편차 / 분산 / 표준편차(l2 norm) / 평균절대편차(l1 norm) / 중간값의 중위절대편차 / 범위 / 순서통계량 / 백분위수 / 사분위범위

<br>

평균절대편차(l1) = $\frac{\sum^n_{i=1}{|x_i - \bar{x}|}}{n}$

분산  = 표준편차(l2) = $\frac{\sum{(x_i - \bar{x})^2}}{n-1}$

여기서 표준편차에 n이 아닌 n-1을 나누는 이유는 만약 표본을 가지고 모집단을 추정한다면 n으로 나눌때 모집단의 참값을 과소평가 하기때문.  
이를 편향 추정이라하고 n-1을 나누면 비평향 추정이 된다.  
"표준편차는 표본의 평균에따른다"라는 제약조건 하나가 있기 때문에 자유로를 n이 아닌 n-1로 둔다.

#### 분포
데이터가 전반적으로 어떻게 분포하고 있는지도 알아봐야함
> Boxplot, Frequency table(도수분포표)과 Histogram, Density plot

#### 기타 데이터 타입
> 최빈값, 기댓값, Bar chart, Pie chart 등

### 3) 상관관계

보통 데이터를 살펴본다고 하면 특별한 목적 (모델링)을 위한 경우이고, 이 목적들은 대부분 AA를 가지고 BB를 CC한다. 로 표현 할 수 있음.  
보통의 문제에선 AA가 데이터에 담긴 여러 Feature들을 의미하고(독립변수), BB 는 모델링 하려는 목적값을 의미 (종속변수), CC는 예측, 추정, 분류 등의 문제들.  
일반적으로 예측값을 X, 목푯값을 Y라고 두는데 이 X와 Y의 관계를 살펴보는것도 EDA과정에서 중요한 과정이다.  

수치형 데이터끼리의 상관관계를 구할때 피어슨 상관계수를 계산한다. 변수 1과 변수 2 각각의 평균으로부터의 편차들을 서로 곱한 값들의 평균을 각 변수의 표준편차의 곱으로 나눠준다.

$r = \frac{\sum^N_{i=1}{(x_i-\bar{x})(y_i-\bar{y})}}{(N-1)s_xs_y}$

상관계수는 -1과 1 사이의 값인데 -1에 가까울수록 음의 상관관계, 1에 가까울수록 양의 상관관계를 의미한다.  
하지만 만약 변수들이 선형적인 관계를 갖지 않는다면, 이는 유용한 지표가 아니다.

데이터의 각 피쳐들에 대해 상관계수를 계산하고 이를 행렬로 나타내면 이를 correlation matrix라 한다.

데이터의 값 보다는 순위를 기초로 하는 상관계수로서 Spearman's rho나 Kendall's tau도 있다.  
피어슨 상관계수는 데이터의 특잇값에 민감한데 값 자체가 아닌 순서에 주목하면서 좀 더 로버스트하고, 비선형 관계의 데이터도 다룰 수 있다.  
하지만 일반적으로 피어슨 아니면 피어슨을 기초로한 로버스트한 버전들을 사용한다고 함.  

상관계수를 그리기 전에 1차적으로 먼저 해볼것은 산점도(scatter plot)로 먼저 간을 한번 봐야한다.

### 4) 다변량분석

위의 분석들은 일변량 분석과 이변량 분석이다. 셋 이상의 변수를 다룰땐 다변량 분석을 사용

수치형끼리는 hexagonal binning, contour plot 그리고 주로 많이 쓰이는게 Heat map 등이 있다.  
범주형 끼리의 비교는 분할표 (contingency table)를 사용한다. (pivot)  
범주형과 수치형끼리의 비교는 Box plot, Violin plot을 사용
